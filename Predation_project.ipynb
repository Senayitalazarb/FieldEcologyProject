## Difference of Predation Rates between Cryptic and Aposematic Strategies!
                                       using artificial caterpillars


### Ecological Background & Hypotheses

Ecologically insect larvae (caterpillars) are an important part of the food chain. In many habitats caterpillars of Lepidoptera order is abundantly found and thus are highly vulnerable to predation. Cryptic, mimicry and aposematism are generally adopted by caterpillars as a defence against predation, under which cryptic and aposematic patterns (AP) are protection against visually detectable predation. The cryptic is related to camouflaging, aposematism signals toxicity clearly discriminating it from the edible cryptic prey and mimicry is the process of morphologically looking like another object or organism. But many times the predators break through these defence strategies.

The cinnabar moth (Tyria jacobaeae L.) which is in the order of Lepidoptera is found naturally across Europe (apart from extreme north) and extends up to Asia. Its larvae feeds mainly on ragwort plant (Jacobaea vulgaris) (RP) and few other Jacobaea species. Due to its more specific feeding nature it is also considered as a successful biological agent to control the problematic RP. They are orange coloured with black stripes. When viewed at a distance they blend together with the background, thus adopting good camouflage along with salient aposematism signals. As a result they are generally rejected by birds but prayed by arthropods. The caterpillar as a group strip off the RP leaves and flower heads but rarely kills the plant. The toxic alkaloids present in the RP is a major problem to the grazing livestock. The cinnabar moth caterpillar (CMC) consumes these toxic alkaloids, stores in their body without digesting and uses as a defence mechanism. 

The effectiveness of the cryptic and aposematic strategy adopted by caterpillars determines their survival, which can be established by studying the pray avoiding behaviour of predators. Thus, estimating the effectiveness of the cryptic and aposematic strategy adopted can explain the status and its continued existence. This can be estimated to a greater accuracy by directly measuring the mortality rates. Using artificial caterpillar (AC) models is a suitable tool for assessing predation by predators like birds, mammals and ants, based on the marks created on them. 

In this study, we use AC in four different areas of Freiburg to test the following hypothesis: (1) The predation rates of the cryptic and aposematic caterpillars are the same (null hypothesis) or significantly different (alternative hypothesis) and (2) The color types of plant where these dummy caterpillars are installed have no effect (null hypothesis) or have a significant effect on the predation (alternative hypothesis)


#### Methodology Adopted
The study was conducted on four randomly selected locations inside Freiburg im Breisgau, Germany based on the abundance of RP in the area (Fig). The sites were, S-1 (Wiehre Bahnhof, 47°58'57.0"N 7°51'20.3"E), S-2 (Littenweiler close the Dreisam, 47°59'34.8"N 7°52'53.7"E), S-3 (Vineyard between St. Georgen and Erbringen, 47°58'16.2"N 7°47'28.5"E) and S-4 (Erbringen close to the main road, 47°57'32.9"N 7°46'10.2"E). The AC were made using non hardening Weible Fantasia plasticine modelling clay, thus satisfying the conditions mentioned by Barber. Two types of AC, pure green (GAC) and yellow with black stripes (YAC) were used in the study. Each caterpillar had a length of 30 mm and 3 mm diameter. A total of 180 AC was made (90 GAC and YAC each). The RP can grow to an height of 175-200 cm, but their usual height ranges between 20-80 cm, thus it was decided to place the AC at an height of 60 cm above ground. To compare the effects of predation, AC was place on RP and any green plant (GP) without flowers, with each plant having one GAC and YAC. The plants with the AC were separated at least by 4 m.


The AC were fixed on the plants using pins, given an identification number, a picture taken using a digital camera with a calibration colour ticket (Fig) for colour analysis and was left on the field for 72 hours. After which, they were collected and examined using a magnifying glass to identify predation marks. The predation marks were accepted only if minimum 3 out of 4 the four examiners accepted and similar multiple marks were considered as a single attack. Based on the marks, the attack was classified as either birds, anthropoids or mammal predation. Chi-square test was used to analyse the correlation between the predator type and the colour of the caterpillars. General Linear Mixed Models (GLMM) was used to test, (1) the two hypothesises of the study, taking colour of the AC and background as fixed effect and the site as random effect (2) the relationship between the quantified colour contrast and the predation risk, taking colour contrast between the AC and background as fixed effect and the site as random effect.


For the colour analysis the RBG vales of the AC and background was obtained using ImageJ® software and transferred into HSV values using matplotlib.colors.rgb_to_hsv in python. The contrast between the colours was done by calculating the Euclidean distance between two colours in an HSV colour space. 




#### Experimental Design!
- Hypothesis 1:  P(predated | aposematic) = P(predated | cryptic)
- Hypothesis 2:  P(predation) ⊥ plant 




This looks like a small data set, where each row represents an observation. Here is a brief description of each column:

- Site: the site where the observation was made
- Plant_id: the identifier for the plant on which the observation was made
- COLOR_caterpillar: the color of the caterpillar observed on the plant
- PREDATION: a binary variable indicating whether or not predation was observed on the plant
- PREDATOR: the number of predators observed on the plant
- CONTRAST: the color contrast between the caterpillar and the plant background
- plant_color: the color of the plant on which the observation was made

> This data set could be used to test hypotheses related to the effect of color and color contrast on predation risk, using statistical methods such as General Linear Mixed Models (GLMM). The site variable could be included as a random effect in the GLMM, to account for the variation between sites. The COLOR_caterpillar and plant_color variables could be included as fixed effects, to test their effect on predation risk. The CONTRAST variable could also be included as a fixed effect, to test the relationship between color contrast and predation risk.

### Model Preparation

### Model 1: to test hypothesis 1 and 2
Hypotheses :

Hypothesis 1: The color of the caterpillar and the color of the plant background significantly influence predation risk on the plant. Specifically, caterpillars with colors that are more similar to the plant background will have lower predation risk.
- add null hyp.

Hypothesis 2: The site where the observation was made has a significant effect on predation risk on the plant, even after controlling for the effect of caterpillar and plant color.

>Model: GLMM with binomial distribution and logit link function.
- Dependent variable: PREDATION
- Fixed effects: COLOR_caterpillar, plant_color, CONTRAST
- Random effect: Site

```python
model1 = smf.mixedlm("PREDATION ~ COLOR_caterpillar + bg", data=df_new, groups= df_new["site"], re_formula="1").fit()
print(model1.summary())
```

Based on the output of the mixed linear model regression, we can interpret the results as follows:

- The intercept of the model is 0.706, which represents the expected value of PREDATION when both COLOR_caterpillar and plant are at their reference levels (yellow and green, respectively).
- The coefficient for COLOR_caterpillar[t. yellow] is -0.303, which means the predation risk of the yellow
    caterpillars always lower, independent of the color of the background plant, we would expect a decrease in the odds of predation by a factor of exp(-0.303) = 0.738, holding all other variables constant.
- The coefficient for bg[T.yellow] is -0.009, which is not statistically significant (P = 0.912). This means that there is no evidence of an effect of the background color on predation risk.
- The estimated variance of the random effect for the site is 0.021. This tells us how much the predicted values for PREDATION can vary across different sites, after accounting for the fixed effects in the model.

Overall, the results suggest that the color of the caterpillar has a significant effect on predation risk, with yellow caterpillars being less likely to be preyed upon than green caterpillars. However, there is no evidence of an effect of the background color on predation risk. The random effect for the site indicates that there is some variation in predation risk that is specific to each site. Moreover, we can hardly assign the yellow or green dummy caterpillars to be either cryptic or aposematic because of what we see from the color analysis

### Model 2: testing the relation between the quantified color contrast and the predation rate

Hypothesis (2): The color contrast between the caterpillar and the plant background is a significant predictor of predation risk on the plant. Specifically, higher color contrast between the caterpillar and the plant background will be associated with higher predation risk.

>Model:GLMM with binomial distribution and logit link function.
- Dependent variable: PREDATION
- Fixed effect: CONTRAST
- Random effect: Site

```python
model2 = smf.mixedlm("PREDATION ~ CONTRAST", data=df_new, groups= df_new["site"], re_formula="1").fit()
print(model2.summary())
```

Based on the output of the mixed linear model regression, we can interpret the results as follows:

- The model suggests that there is a significant positive relationship between color contrast and predation risk (p < 0.01). The intercept of 0.348 suggests that the predicted probability of predation when the color contrast is zero is 0.348. The coefficient for CONTRAST (0.297) suggests that, for a one-unit increase in contrast, the predicted probability of predation increases by 0.297. The random effect variance for SITE is estimated at 0.022, indicating that there is some variation in predation risk across sites that is not explained by the fixed effect of color contrast.

- It is important to note that this model assumes a linear relationship between color contrast and predation risk. It is possible that the relationship is more complex and may involve interactions with other variables. Further exploration and analysis may be necessary to fully understand the relationship between color contrast and predation risk in this system.

### Checking variation in the response variable

```python
# Extract random effects
random_effects = pd.DataFrame(model1.random_effects)
re = pd.DataFrame(random_effects)
site_rf = re.T
site_rf = site_rf.reset_index().rename(columns = {'index': "site"})
site_rf
```

- model1.random_effects is used to extract the random effects of the mixed effects model. In this case, since the site was treated as a random effect, the random effects extracted are the deviations of each site from the overall intercept.


```python
# Extract residual variance
resid_var = model1.scale
print('Residual_variance:', resid_var)
```

- model1.scale is used to extract the residual variance of the mixed effects model. This represents the amount of unexplained variation in the response variable that is not accounted for by the fixed or random effects.

```python
import matplotlib.pyplot as plt

residuals = model1.resid
fitted_values = model1.fittedvalues

plt.scatter(fitted_values, residuals)
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted values plot")
plt.show()
```

### Plot_Summary
- Plot of the residuals (difference between the actual values and predicted values) versus the predicted values (fitted values) from the GLMM model. 
- This plot is useful for checking if the model assumptions are met, particularly the assumption of constant variance of the residuals across the range of the predictor variable. If the plot shows a random pattern with no clear trend or pattern, then the model assumptions are likely met. 
- However, if there is a clear pattern (such as a funnel shape), then the model may not be appropriate for the data and further investigation is needed.
